```python
from ultralytics import YOLO
```

```python
# Load a model

model = YOLO('yolov8m.pt')  # load a pretrained model (recommended for training)
```

```python
pwd
```

```python

# Train the model
model.train(data='/content/gdrive/MyDrive/Project-KneeXraysSegmentation/custom-yolo-roboflow/knee-dataset-yolov8-format/data.yaml',
            epochs=250,
            imgsz=640,
            val=True,
            patience=50,
            batch=32,
            save=True,
            optimizer="SGD")
```
