```python
import openpyxl
import cv2
```

```python
# excel_filename where excel file and actual X-ray images located
excel_filename = "evaluation_parameter.xlsx"
actual_image_path = "/home/keva/Documents/KneeSegmentationProject/Data/KneeXray/ActualImage/"
```

```python
workbook = openpyxl.load_workbook(filename=excel_filename)
workbook.sheetnames
```

```python
# duplicating the original sheet.
# u r free to do changes in duplicated sheet. sheet name is "Sheet1 Copy"
sheet = workbook.active
workbook.copy_worksheet(sheet)
workbook.sheetnames
```

```python
# load the duplicate sheet
sheet = workbook["Sheet1 Copy"]
```

```python
# glance the head column names
for row in sheet.iter_rows(min_row=1,
                           max_row=1,
                           values_only=True):
    print(row)
```

```python
image_names = [name.value for name in sheet["A"][1:]]
print(*image_names[0:6], *[".", ".", "."], *image_names[-5:],f"total Knee-Xray: {len(image_names)}", sep="\n")
```

```python
sheet["C1"] = "USABLE IMAGES"
USABLE_IMAGES = "C"
# Image window dimensions
down_points = (1280, 720)
```

```python
def mark(row, column, sheet, action):
    loc = column + str(row)
    sheet[loc] = 0 if action == 120 else 1 if action == 99 else "Invalid input"
    return
```

```python
for idx, name in enumerate(image_names):
    image_path = actual_image_path + name
    image = cv2.imread(image_path, 0)
    resized_down = cv2.resize(image, down_points, interpolation= cv2.INTER_LINEAR)
    cv2.imshow(name, resized_down)
    action = cv2.waitKey(0)
    mark(idx + 2, USABLE_IMAGES, sheet, action)
    cv2.destroyAllWindows()
    workbook.save(filename=excel_filename)
    print(idx+2, "completed")
```
