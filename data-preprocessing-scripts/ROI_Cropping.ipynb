    Script open up the images listed in images_path folder in a window called "Knee X-rays",

    u can crop the image by pressing down left button in mouse and dray the area you desired,
    then by releasing the left button.

    By pressing down any key *prefer space bar. It goes in next images in order of os.listdir().

    It save the cropped image in the target_path folder.
    Name of saved image is same as source image.


```python
import os
import cv2 as cv
```

```python
images_path = "/home/keva/Documents/KneeSegmentationProject/Data/KneeXray/UsableImages/"
target_path = "/home/keva/Documents/KneeSegmentationProject/Data/KneeXray/ResizedImageSorted/"
```

```python
down_points = (840, 840)
output_size = (448, 448)
```

```python
images_name = os.listdir(images_path)
```

```python
window_name = "Knee X-rays"
cv.namedWindow(window_name)
```

```python
global resized_down, target_image_path, cropping
```

```python
def crop_image(event, x, y, flags, param):
    global x_start, y_start, x_end, y_end, cropping

    if event == cv.EVENT_LBUTTONDOWN:
        x_start, y_start, x_end, y_end = x, y, x, y
        cropping = True

    elif event == cv.EVENT_MOUSEMOVE:
        if cropping:
            img2 = resized_down.copy()
            x_end, y_end = x, y
            cv.rectangle(img2,
                     (x_start, y_start),
                     (x_end, y_end),
                     (255, 255, 255),
                     1)
            cv.imshow(window_name, img2)

    elif event == cv.EVENT_LBUTTONUP:
        x_end, y_end = x, y
        crop_img = resized_down[y_start: y_end, x_start:x_end]
        resized_image = cv.resize(crop_img, output_size, interpolation= cv.INTER_LINEAR)
        cv.imwrite(target_image_path, resized_image)
        cropping = False

cv.setMouseCallback(window_name, crop_image)
```

```python
cropping = False

for idx, name in enumerate(images_name):
    image_path = images_path + name
    target_image_path = target_path + name

    image = cv.imread(image_path, 0)
    resized_down = cv.resize(image, down_points, interpolation= cv.INTER_LINEAR)

    cv.imshow(window_name, resized_down)

    action = cv.waitKey(0)

    if action == 113:
        break

    print(f"Image name: {name}, "
          f"image index: {idx}, "
          f"total {idx + 1} completed",
          sep="\n")

cv.destroyAllWindows()
```
